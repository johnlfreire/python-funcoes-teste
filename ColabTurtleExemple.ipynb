```
# Importando o modulo
import ColabTurtle.Turtle as turtle
# Vamos usar o math mais a frente
import math

# 
# Função que desenha um círculo com a ColabTurtle
#
# Entradas : t = objeto turtle.
#            r = raio da círculo (int).
#            color = cor do retângulo (String) em turtle.VALID_COLORS.
#
def circle(t, r, color="red"):
    t.color(color)
    perimeter = 2 * math.pi * r        
    n = int(perimeter / 3) + 3
    step = int(perimeter / n)
    angle = 360/n
    for i in range(n):
        t.forward(step)
        t.left(angle)

def simbolo_olimpiadas(t, r, xc, yc, color="red"):  
    t.bgcolor("white")
    # Círculo azul
    t.penup()
    t.goto(xc, yc)

    # Retirando a caneta
    t.pendown()
    circle(t, r, color)
    
    t.penup() 
    t.goto(xc+150, yc)
    
    # Círculo preto
    
r = 100
width = 5 #@param {type:"integer"}
speed = 10 #@param {type:"slider", min:1, max:10, step:1}

# Inicializando a turtle
turtle.initializeTurtle()
turtle.speed(speed)
turtle.width(width)

# Posição inicial
xc = 200
yc = 200
simbolo_olimpiadas(turtle, r, xc, yc, color="blue")
#turtle.hideturtle()
xc = 280
yc = 280
simbolo_olimpiadas(turtle, r, xc, yc, color="yellow")
xc = 360
yc = 200
simbolo_olimpiadas(turtle, r, xc, yc, color="black")
turtle.hideturtle()
xc = 440
yc = 280
simbolo_olimpiadas(turtle, r, xc, yc, color="green")
turtle.hideturtle()  
xc = 520
yc = 200
simbolo_olimpiadas(turtle, r, xc, yc)
turtle.hideturtle()  
```

```
def rectangle(t, b, h, color="red"):
    t.color(color)
    t.right(90)
    t.forward(b)
    t.right(90)
    t.forward(h)
    t.right(90)
    t.forward(b)
    t.right(90)
    t.forward(h)
    
def quadrados_con(t, xc, yc):
    #Cor de fundo
    t.bgcolor("white")
    # Cores válidas
    colors = turtle.VALID_COLORS

    # Primeiro retângulo
    t.penup()
    t.goto(xc+50, yc+50)

    # Retirando a caneta
    t.pendown()
    rectangle(t, 100, 100, color=colors[1])
    

    # Segundo retângulo...
    t.penup()
    t.goto(xc+25, yc+25)
    
    t.pendown()
    rectangle(t, 150, 150, color=colors[2])

    # Segundo retângulo...
    t.penup()
    t.goto(xc+0, yc+0)
    
    t.pendown()
    rectangle(t, 200, 200, color=colors[3])
    
    # Segundo retângulo...
    t.penup()
    t.goto(xc-25, yc-25)
    
    t.pendown()
    rectangle(t, 250, 250, color=colors[4])
    
    # Segundo retângulo...
    t.penup()
    t.goto(xc-50, yc-50)
    
    t.pendown()
    rectangle(t, 300, 300, color=colors[5])
    
    
speed = 10 #@param {type:"slider", min:1, max:10, step:1}
width = 3 #@param {type:"integer"}

# Inicializando a turtle
turtle.initializeTurtle()
turtle.speed(speed)
turtle.width(width)

# Posição inicial
xc = 250
yc = 150

# Desenhando...
quadrados_con(turtle, xc, yc)
turtle.hideturtle()
```

```
# 
# Função que desenha um triângulo com a ColabTurtle
#
# Entradas : t = objeto turtle.
#            a = um lado do triângulo (int).
#            b = um lado do triângulo (int).
#            c = um lado do triângulo (int).
#            color = cor do retângulo (String) em turtle.VALID_COLORS.
#
#            Observação: |b-c|<a<b+c
#
def triangle(t, a, b, c, color="red"):
    
    t.color(color)
    
    if not(abs(b-c)<a<b+c):
        raise ValueError('Não é possível desenhar um triângulo com esses lados')
    
    turn = lambda a,b,c: math.degrees(math.pi - math.acos((a*a+b*b-c*c)/(2*a*b)))
    t.forward(a)
    t.left(turn(a,b,c))
    t.forward(b)
    t.left(turn(b,c,a))
    t.forward(c)  
def estrela_seis_pontas(t, a, b, c, xc, yc):
    
    # Posição inicial
    turtle.penup()
    turtle.goto(xc, yc)
    turtle.pendown()
    
    # Primeiro triângulo
    t.left(270)
    triangle(t, a, b, c)
    
    # Primeiro triângulo
    t.left(300)
    triangle(t, a, b, c)
    
    for i in range(4):
      t.left(180)
      t.forward(a)
      triangle(t, a, b, c)
     
a = 80 #@param {type:"integer"}
b = a
c = a 
width = 3 #@param {type:"integer"}
speed = 5 #@param {type:"slider", min:1, max:10, step:1}

# Inicializando a turtle
turtle.initializeTurtle()
turtle.speed(speed)
turtle.width(width)

# Posição inicial
xc = 350
yc = 200
estrela_seis_pontas(turtle, a, b, c, xc, yc)
turtle.hideturtle()
```
